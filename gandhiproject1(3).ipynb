```python
#encode the dependent variable containing categorical values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
encoder.fit(y)
```

```python
y = encoder.transform(y)
```

```python
y
```

```python
#transform the data in training and testing
from sklearn.utils import shuffle
X,y = shuffle(X,y,random_state=1)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
train_x =sc.fit_transform(train_x)
test_x = sc.transform(test_x)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
keras.__version__
```

```python
model=Sequential()
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
model.add(Dense(input_dim=60,init="random_uniform",activation='relu',output_dim=15))#input layer
```

```python
model.add(Dense(output_dim=8,init="random_uniform",activation='relu'))#hidden layer
```

```python
model.add(Dense(output_dim=1,init='random_uniform',activation='sigmoid'))#output layer
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])#adam=batch gradent descent
```

```python
#train_y.reshape(1,-1)
train_y.shape
```

```python
model.fit(train_x,train_y,epochs=150, batch_size=83)#epochs no of iteration 
```

```python
y_pred = model.predict(test_x)
```

```python
y_pred=(y_pred>0.5)
```
